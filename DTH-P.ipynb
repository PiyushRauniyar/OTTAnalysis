```python
def clean_genre_list(list_):       #defining library to slice plot description
    if type(list_) is str:
        list_ = "[" + list_ + "]"
        list_ = list_.replace(',', '","')
        list_ = list_.replace('[', '["')
        list_ = list_.replace(']', '"]')
        list_ = list_.replace(' "', '"')
        list_ = list_.replace(" ","")
        return list_
    else:
        return "[]"
```

```python
dth["genres"] = dth["genres"].apply(clean_genre_list)
dth["genres"] = dth["genres"].apply(eval)
```

```python
def to_1D(series):
    return pd.Series([x for _list in series for x in _list])
```

```python
df_genre_imdb = dth.groupby(to_1D(dth["genres"])).mean()
print(df_genre_imdb.sort_values(["Ratings"],ascending=False))
```

```python
df_genre_count = to_1D(dth["genres"]).value_counts()
print(df_genre_count)

df_genre_count.plot(kind = 'bar')
plt.show()
```

```python
def findstart(list_):
    if type(list_) is str:
        list_ = list_[:4]
        return list_
    else:
        return ""

def duration(list_):
    if type(list_) is str:
        if len(list_) == 9 and list_[0]!="I":
            l1 = int(list_[:4])
            l2 = int(list_[5:])
            return l2-l1
        else:
            return 0
    else:
        return 0

dth["First Year"] = dth["Year"].apply(findstart)
dth["Years Run"] = dth["Year"].apply(duration)        
```

```python
df_year_count = dth["First Year"].value_counts().sort_index()
df_year_count = df_year_count.iloc[:-4] #removing the weird I values

df_year_count.plot(kind = 'bar')
plt.show()
```

```python
print(dth.sort_values(["Years Run"], ascending=False))
```

```python
def clean_ovw_list(list_):

    if type(list_) is str:
        list_ = "[" + list_ + "]"

        #removing all the special characters
        list_ = list_.replace(',', '')
        list_ = list_.replace('.', '')
        list_ = list_.replace('"', '')
        list_ = list_.replace('(', '')
        list_ = list_.replace(')', '')
        list_ = list_.replace('-', '')
        list_ = list_.replace('»', '')

        list_ = list_.replace(' ', '","')
        list_ = list_.replace('[', '["')
        list_ = list_.replace(']', '"]')
        list_ = list_.replace(' "', '"')

        #converting to lower case
        list_ = list_.lower()
        return list_
    else:
        return "[]"
```

```python
dth["overview"] = dth["overview"].apply(clean_ovw_list)
dth["overview"] = dth["overview"].apply(eval)
```

```python
df_ovwcount = dth.groupby(['First Year',to_1D(dth["overview"])]).count().reset_index()
```
